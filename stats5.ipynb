```python
# nort  正規分布
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

x = np.linspace(-3,3,100)
y = stats.norm.pdf(x)
plt.plot(x,y)
```

```python
#CDF  累積

x = np.linspace(-3,3,100)
y = stats.norm.cdf(x)
plt.plot(x,y)
```

```python
#正規分布  scale = 標準偏差　ばらつき　loc = 平均
x = np.linspace(-5, 10, 100)

y = stats.norm(loc=2, scale=3).pdf(x)
y2 = stats.norm(loc=3, scale=3).pdf(x)

plt.plot(x,y)
plt.plot(x,y2)
```

```python
#カーネル密度推定
x = np.linspace(0.10,10)
y = stats.gaussian_kde([1,1,2,2,3,4,5]).pdf(x)
plt.plot(x,y)
```
