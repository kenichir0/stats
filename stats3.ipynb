```python
#  連続型
def simple_func(x):
    return 2 * x +3


x =np.linspace(-3,3,100)
y = simple_func(x)
```

```python
#  離散型
def simple_func(x):
    return x**2 + 3


x =np.arange(-3,4)
y = simple_func(x)
plt.plot(x,y)
```

```python
x = np.linspace(-2,3,100)
y = stats.norm.cdf(x)
```

```python
plt.plot(x,y)
```

```python
x = np.linspace(-2,3,100)
y = stats.norm.pdf(x)
```

```python
plt.plot(x,y)
```

```python
x = np.linspace(-5,5, 100)
y = stats.norm(loc=10, scale=3).pdf(x)
y1 = stats.norm(loc=1, scale=3).pdf(x)
```

```python
plt.plot(x,y)
plt.plot(x,y1)
```

```python
x = np.linspace(-5,5, 100)
y = stats.norm(loc=10, scale=3).cdf(x)
y1 = stats.norm(loc=1, scale=3).cdf(x)
```

```python
plt.plot(x,y)
plt.plot(x,y1)
```
